```python
#Setting up notebook

%matplotlib notebook
import matplotlib.pyplot as plt

# Make plots live-update while scans run.
from bluesky.utils import install_nb_kicker
install_nb_kicker()

from bluesky.plans import count
from bluesky.examples import noisy_det, det, motor, det1, det2
from bluesky.callbacks import LivePlot, LiveTable
from bluesky import RunEngine
```

```python
RE = RunEngine({})
```

```python
RE.waiting_hook = ProgressBarManager()
```

### Multiple Plot steps
- Initialize plot
- setup plot "variable" (in order to keep RE call manageable)
- RE call

```python
#"initialize plot":

fig, ax = plt.subplots()
```

```python
#Setup plot variable as list of LivePlot objects

#I think the plots can easily accommodate more than two plots by adding extra LivePlot calls.
#Change noisy_det and det2 to the correct signal names matching those in the count call in the next frame

multiLivePlot = [LivePlot(noisy_det, ax = ax, label='noisy_det'),LivePlot(det2, ax = ax, label='det2'),LivePlot(det1, ax = ax, label='det1')]
```

```python
#RunEngine call

#Starting count and updating plot. simply change noisy_det and det2 to the correct signal names

RE(count([det1, noisy_det, det2], delay = 3, num = None), multiLivePlot)
```
